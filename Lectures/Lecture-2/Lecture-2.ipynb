### Reverse a vector

1 4 2 6
1 2 3 4

6 2 4 1


```r
reverse_vector <- function(v) {
  rv = rep(0, length(v))
  for (i in 1:length(v)) {
    rv[length(v) - i + 1] <- v[i]
  }
  return(rv)
}
v <- c(1, 4, 2, 6)
cat(reverse_vector(v))
```

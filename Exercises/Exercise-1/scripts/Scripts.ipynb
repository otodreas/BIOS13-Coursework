### Random walk plot

```r
walk_length <- readline("Input random walk length: ")
walk_length <- as.numeric(walk_length)

y <- 0
for (i in 1:(walk_length-1)) {
  y[i+1] <- y[i] + rnorm(1)
}

plot(y, type="l")
```

### Vector manipulator
Take 2 vectors of length 3, alternatingly take values from each vector in the new

```r
v1 <- c(-7, -8, -9)
v2 <- c(40, 50, 60)
v3 <- 1:6

for (i in 1:length(v1)) {
  v3[i*2-1] <- v1[i]
  v3[i*2] <- v2[i]
}

cat(v3)
```

### Vowel finder
print the vowels in an input (text string)

```r
string <- readline("Input phrase: ")
vowels <- c("a", "e", "i", "o", "u", "å", "ä", "ö")

cat("Vowels:")

i <- 0
repeat {
  i <-  i + 1
  letter <- substr(string, i, i)
  if (letter != "") {
    if (letter %in% vowels) {
      cat("", letter)
    }
  } else {
    break
  }
}
```

### Guessing game
- Randomly generate a number 1:20
- Have a user guess the number
- Tell them if theyre too low or too high
- when correct, print the number of guesses

```r
secret_number <- sample(1:20, 1)

guess <- 0
i <- 0

while (guess != secret_number) {
  i <- i + 1
  guess <- as.integer(readline("Guess the number: "))
  if (guess < secret_number) {
    cat("Wrong: guess too small.\n")
  } else if (guess > secret_number) {
    cat("Wrong: guess too large.\n")
  }
}

cat("Correct guess after", i, "guesses.")
```

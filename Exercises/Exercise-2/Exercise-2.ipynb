### Question 6
Create a function that takes two lists and combines them and sorts them

```r
# Define function
sort_multi_vectors <- function(v1, v2) {
  # Create a vector containing both input vectors
  v3 <- c(v1, v2)
  # Initialize v3_trim to iteratively cut off
  v3_trim <- v3
  # Loop through the length of the two input vectors
  for (i in seq_along(v3)) {
    # Get the index of the element to trim off the vector v3_trim
    trim_element <- length(v3)-i+1
    # Trim everything following the index of trim_element
    v3_trim <- v3_trim[1:(trim_element)]
    # Loop through indices in v3_trim, omitting the last
    for (j in seq_along(v3_trim[-length(v3_trim)])) {
      # Check if pair of values are not sorted
      if (v3_trim[j] > v3_trim[j+1]) {
        # Reverse pair
        v3_trim[(j+1):j] <- v3_trim[j:(j+1)]
      }
      # After bubble sort iteration, update the final element of v3_trim
      v3[trim_element] <- v3_trim[trim_element]
    }
  }
  # Return sorted vector v3
  return(v3)
}

v1 <- c(2, 8, 30)
v2 <- c(6, 5)
cat(sort_multi_vectors(v1, v2))
```

### Question 4
Plotmax function: plot random vector of length n, highlight local maxima with red cross

```r
# Create plotting function
plotmax <- function(n) {
  # Generate series
  random_series <- runif(n)
  # Plot series
  plot(random_series, type="l")
  # Initialize skip
  skip <- F
  # Loop through series, omit last two indices
  for (i in seq_along(random_series[1:(n-2)])) {
    # Check if the previous iteration contained a maximum
    if (skip) {
      # Reinitialize skip to False
      skip <- F
      # Skip this iteration since there cannot be a local maximum
      next
    }
    # Create subvector of length 3
    subv <- random_series[i:(i+2)]
    # Check if subvector contains local maximum
    if (subv[2] > subv[1] && subv[2] > subv[3]) {
      # Plot the maximum
      points((i+1), subv[2], pch=4, col="red")
      # Assign True to skip to ignore the next iteration
      skip <- T
    }
  }
}
# Run function
plotmax(20)
```
